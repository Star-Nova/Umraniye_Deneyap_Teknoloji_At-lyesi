Python'da çeşitli veri tipleri bulunur ve her biri farklı türde veri depolamak için kullanılır. İşte Python'un temel veri tipleri ve bunların açıklamaları:

### Sayısal Tipler
1. **integer (int)**: Tam sayılar için kullanılır. Örnek: `5`, `-3`, `42`
2. **float (float)**: Ondalıklı sayılar için kullanılır. Örnek: `3.14`, `-0.001`, `2.718`
3. **complex (karmaşık sayı)**: Gerçek ve sanal kısmı olan sayılar için kullanılır. Örnek: `3 + 4j`, `1 - 2j`

### Metin Tipi
1. **string (str)**: Metin verilerini tutar. Çift tırnak veya tek tırnak içinde yazılır. Örnek: `"merhaba"`, `'dünya'`

### Koleksiyon Tipleri
1. **list (liste)**: Sıralı ve değiştirilebilir veri koleksiyonlarıdır. Köşeli parantez içinde yazılır. Örnek: `[1, 2, 3]`, `['elma', 'muz', 'çilek']`
2. **tuple (demet)**: Sıralı ama değiştirilemez veri koleksiyonlarıdır. Parantez içinde yazılır. Örnek: `(1, 2, 3)`, `('elma', 'muz', 'çilek')`
3. **set (küme)**: Sırasız ve eşsiz (unique) veri koleksiyonlarıdır. Süslü parantez içinde yazılır. Örnek: `{1, 2, 3}`, `{'elma', 'muz', 'çilek'}`
4. **dictionary (sözlük)**: Anahtar-değer çiftleri koleksiyonlarıdır. Süslü parantez içinde yazılır ve her çift iki nokta üst üste ile ayrılır. Örnek: `{'isim': 'Ahmet', 'yaş': 25}`, `{'elma': 3, 'muz': 5}`

### Boolean Tipi
1. **bool (boolean)**: `True` veya `False` değerlerini tutar. Mantıksal ifadeler için kullanılır. Örnek: `True`, `False`

## 2. Ders
**Makine öğrenmesi (Machine Learning)**, bilgisayarların veri kullanarak öğrenme ve belirli görevleri otomatik olarak gerçekleştirme yeteneğini kazanmasını sağlayan bir yapay zeka (AI) dalıdır. Bu, sistemlerin, önceden programlanmış açık talimatlar olmadan, verilerden örüntüler ve içgörüler çıkarmalarını sağlar. Makine öğrenmesi, veri analizinde, model tahmininde ve karar verme süreçlerinde yaygın olarak kullanılır.

### Makine Öğrenmesinin Temel Bileşenleri

1. **Veri**: Makine öğrenmesi modellerinin eğitilmesi için kullanılan ham bilgiler. Bu, sayısal veriler, metinler, görüntüler veya diğer veri türleri olabilir.
2. **Özellikler (Features)**: Modellerin öğrenmesi ve tahmin yapması için kullanılan veri noktalarının özellikleridir. Örneğin, bir evin fiyatını tahmin etmek için evin metrekaresi, oda sayısı, lokasyonu gibi özellikler kullanılır.
3. **Model**: Verileri analiz eden ve tahminler veya kararlar veren matematiksel yapı. Modeller, belirli bir görevi yerine getirmek üzere eğitilir.
4. **Algoritma**: Modeli eğitmek ve güncellemek için kullanılan yöntemlerdir. Algoritmalar, verilerden öğrenme sürecini yönlendirir.
5. **Hedef Değişken (Target Variable)**: Tahmin edilmek istenen sonuç veya değer. Bu, denetimli öğrenmede bilinen bir değerdir.

### Makine Öğrenmesi Türleri

1. **Denetimli Öğrenme (Supervised Learning)**: Modelin, etiketli veri kullanarak eğitildiği öğrenme türüdür. Girdi ve çıktı verileri bilinir ve model, doğru sonuçları tahmin etmek için eğitilir.
    - **Regresyon (Regression)**: Sürekli sayısal değerleri tahmin etmek için kullanılır. Örneğin, ev fiyatı tahmini.Diyabet veri seti.
    - **Sınıflandırma (Classification)**: Verileri kategorilere ayırmak için kullanılır. Örneğin, e-postaların spam olup olmadığını belirlemek.Suyun sicaligini degerlendirirken sicak,ilk veya soguk kategorileri kullanilir.Siniflandirma isleminde kategorik veriler uzerinde yani sozel veriler uzerinde calisilir. Iris veriseti.


### 1. **E-posta Spam Filtreleme:**
   - **Problem:** Bir e-postanın spam (istenmeyen) veya spam olmayan (istenen) olarak sınıflandırılması.
   - **Girdi Verileri:** E-postanın metni, konu başlığı, gönderici adresi, içerik uzunluğu.
   - **Etiketler:** Spam, spam değil.

### 2. **Hastalık Teşhisi:**
   - **Problem:** Bir hastanın belirli bir hastalığa sahip olup olmadığının teşhis edilmesi.
   - **Girdi Verileri:** Hastanın yaş, cinsiyet, kan basıncı, kolesterol düzeyi gibi tıbbi verileri.
   - **Etiketler:** Hastalık var, hastalık yok.

### 3. **Müşteri Terk Tahmini:**
   - **Problem:** Bir müşterinin hizmeti bırakıp bırakmayacağının tahmin edilmesi.
   - **Girdi Verileri:** Müşteri yaşı, cinsiyeti, hizmet kullanım süresi, aylık harcamalar.
   - **Etiketler:** Terk edecek, terk etmeyecek.

### 4. **Kredi Riski Değerlendirme:**
   - **Problem:** Bir başvurucunun kredi alıp alamayacağının tahmin edilmesi.
   - **Girdi Verileri:** Başvurucunun gelir düzeyi, iş durumu, kredi geçmişi.
   - **Etiketler:** Kredi verilir, kredi verilmez.

### 5. **Görüntü Tanıma:**
   - **Problem:** Görüntüdeki nesnenin sınıflandırılması (örneğin, bir resimdeki hayvanın köpek mi kedi mi olduğunu belirlemek).
   - **Girdi Verileri:** Görüntü pikselleri veya özellikleri.
   - **Etiketler:** Köpek, kedi.

### Örnek Senaryo: E-posta Spam Filtreleme

Bir e-posta sağlayıcısı, kullanıcıların gelen kutusuna gelen e-postaların spam olup olmadığını belirlemek istiyor. İşte bu problemi çözmek için denetimli öğrenme sürecinin adımları:

1. **Veri Toplama:** Kullanıcıların e-postaları ve bu e-postaların spam olup olmadığına dair etiketler toplanır.
2. **Özellik Çıkarımı:** E-postaların özellikleri çıkarılır. Bu özellikler e-postanın içeriği, konu başlığı, gönderici adresi, kelime sıklıkları gibi bilgiler olabilir.
3. **Model Eğitimi:** Çıkarılan özellikler ve etiketler kullanılarak bir makine öğrenmesi modeli eğitilir. Örneğin, Naive Bayes, Destek Vektör Makineleri (SVM) veya Lojistik Regresyon gibi algoritmalar kullanılabilir.
4. **Model Değerlendirme:** Modelin performansı, doğruluk, hassasiyet, geri çağırma gibi metriklerle değerlendirilir.
5. **Tahmin:** Eğitilen model, yeni gelen e-postaların spam olup olmadığını tahmin etmek için kullanılır.

Bu şekilde, denetimli öğrenme kullanılarak e-postaların spam olup olmadığını belirlemek mümkün olur ve kullanıcılar istenmeyen e-postalardan korunmuş olur.

2. **Denetimsiz Öğrenme (Unsupervised Learning)**: Modelin, etiketlenmemiş veri kullanarak örüntüleri ve ilişkileri öğrenmeye çalıştığı öğrenme türüdür.
    - **Kümeleme (Clustering)**: Verileri benzerliklerine göre gruplamak. Örneğin, müşteri segmentasyonu.

   - **Yaşam Tarzı:** Aktif, evde vakit geçirmeyi seven, seyahat etmeyi seven

   - **Satın Alma Nedenleri:** Kaliteye odaklananlar, fiyata odaklananlar, hız ve kolaylık arayanlar

Bir online giyim mağazası, müşteri segmentasyonu yapmak istiyor. İşte bazı olası segmentasyon kriterleri:

   - **Alışveriş Alışkanlıkları:** Sık alışveriş yapanlar, sadece özel günlerde alışveriş yapanlar
   - **Kullanım Oranı:** Yeni kullanıcılar, düzenli müşteriler
   - **Satın Alma Nedenleri:** İndirimler ve kampanyalar nedeniyle alışveriş yapanlar, belirli bir markaya sadık olanlar

Bu segmentasyon sayesinde online giyim mağazası, her bir müşteri grubuna özel pazarlama stratejileri geliştirebilir, kampanyalarını daha etkili hale getirebilir ve müşteri memnuniyetini artırabilir.

**Veri Kavramı**  
`Veri kavramı` genellikle durumlar ya da olaylara dair çıkarımlar oluşturmak adına kullanılan
bilgi olarak ifade edilebilir. Bununla birlikte bir başka alternatif bakış açısıyla; bilginin dijital
sistemlerde temsil edilen hali olarak da kabul edilebilmektedir. Veri özellikle günümüz
teknolojisinde önemli bir yer tutmakla beraber, algoritmalarda ve Yapay Zeka gibi gelişmiş
problem çözümleyici yapılarda olmazsa olmaz bir unsur olarak düşünülmektedir. Gününüz
şartları altında teknolojinin ilerleyişi verinin önemini artırmakta ve yine teknolojiyle birlikte
ortaya çıkan yeni problemlere yönelik çözüm araçlarında sebep-sonuç ilişkilerini kurmakta
kullanılmaktadır. Bu nedenle veriler çeşitli araçlarla toplanmakta, geleneksel araçlarla
toplanmış bilgiler verilere dönüştürülmekte ve toplanan ya da işlenen veriler dijital ortamlarda
saklanmaktadır.
Bilgisayar ve yazılım teknolojilerinde meydana gelen ilerlemeler, verilerin dijital ortamda
tutulma şekillerini de etkilemiştir. Bilgisayar sistemlerinin çözmeye çalıştığı problemlerin
karmaşıklaşması, yeni çıkan dijital teknolojilerinin veri üretimin artışında rol oynaması,
bilginin veri düzeyinde nasıl depolanacağı ve nasıl işleneceği gibi soruların güncelliğini
korumasına sebep olmuştur. Buna göre verinin temsil ediliş şekli genel hatlarıyla şu
yaklaşımlar üzerinden ilerlemiştir (Silahtaroğlu, 2008; Provost, & Fawcett, 2013; Larose, &
Larose, 2014):

**Soru:** Veri toplama yontemleri nelerdir ne gibi veriler toplanabilir. Ornegin telefon, akilli saat gibi...

`• Dosya:` Dosyalar ortak nitelikte ve işlevdeki verilerin bir arada tutulup işlenmesini
sağlayan veri araçları olarak ortaya çıkmıştır. Dosya yapılanması halen yaygınlığını
korumakta, farklı türden verilerin kullanıcı nezdinde organize edilmesi ve yine farklı
yazılımlar tarafından kullanılacak şekilde depolanması, dosya adı verilen yapılar
aracılığıyla gerçekleşmektedir. Dosyalar işletim sistemlerinin merkezi veri depolama
unsurları olarak da kabul edilmektedir.

`• Veritabanı:` Fazla miktardaki veriye erişim ve bu verilerin işlenmesi çok sayıda
dosyanın kullanımı sayesinde verimli olmamış; bu nedenle tutulacak verilerin parçalı
yapılarda ve hatta ilişkisel olarak saklanıp kullanılabileceği veritabanı yapıları
geliştirilmiştir. Veritabanları, arkaplanda organize bir biçimde tutulması gereken ve
aktif bir biçimde veri oluşturulması ya da işlenmesi süreçleriyle çalışan yazılımların ve
çevrimiçi platformların olmazsa olmaz unsurları arasında yer almaktadır.  

`• Veriambarı:` Veriambarları, temel işlevi sadece veri depolama yönünde ağırlık
kazanan veritabanlarının bir bakıma veri analizi yetenekleri kazanmış hali olarak
tasarlanmışlardır.

`• Veri Madenciliği Süreçleri ve Büyük Veri:` Günümüzde elde tutulan veriden yeni
bilgi elde edebilmek adına Makine Öğrenmesi ile de ortak paydada buluşan Veri
Madenciliği (Data Mining) çözümleri yaygın bir biçimde kullanılmaktadır. Bazı
Makine Öğrenmesi algoritmaları aynı zamanda Veri Madenciliği algoritması olarak da
kabul edilmektedir. Bu tür araçlar yeni bilgi keşfi için etkin çözümler ortaya koymakta
ve hatta günümüzde gerçek zamanlı olarak, farklı kaynaklardan, yüksek miktarda,
değerli verilerin oluşturduğu ve ileri düzey veri işleme adımlarına ihtiyaç duyan Büyük
Veri (Big Data) için kullanılmaktadır. Günümüzde Yapay Zeka (özellikle Makine Öğrenmesi / Derin Öğrenme) ve Veri Madenciliği Büyük Veri uygulamalarının odak noktası konumundadır.
Verilerin Yapay Zeka gibi ileri düzey uygulamalara hazır hale getirilmesi günümüzde Veri
Bilimi (Data Science) alanının çalışmaları içerisine girmektedir. Veri Bilimi çalışmaları
verilerin etkili bir biçimde işlenebilmesi için mümkün olan en temel veri depolama
yaklaşımlarına başvurmaktadır. Bu amaçla verilerin parametreler eşliğinde tutulabilmesini
sağlayan dosya türleri, “veri seti” kavramının ortaya çıkışında etkili olmuştur. Yapay Zeka
uygulamalarında sıklıkla dile getirilen veri seti, aslında Makine Öğrenmesi tekniklerinin eğitim
ve test gibi aşamalarda kullanabileceği veri bütünlerine karşılık gelmektedir (Arslan, 2019;
Provost, & Fawcett, 2013)


**Veri Seti Kavramı**  
Veri seti özellikle Makine Öğrenmesi algoritmaları üzerinden oluşturulan modellerin hedef
problem için eğitilmesi adına organize bir biçimde bir araya getirilen örnekler bütünü olarak
ifade edilebilmektedir. Buna göre tipik bir veri seti, hedef problemin sebep ve sonuçlarını
(girdiler ve çıktılarını) dikkate alarak (Bkz. 1. Hafta – 1. Gün – 1. Ders ve 1. Hafta – 2. Gün –
2. Ders) eğitilmiş Makine Öğrenmesi modellerini elde etmek için kullanılabilecek veri
bütünlerine tekabül eder (Banachewicz, & Massaron, 2022; Larose, & Larose, 2014). Çok daha
pratik bir bakış açısıyle veri seti, satırlar yönünde örnekler ve sütunlar yönünde parametrelerin
yer aldığı bir Excel tablosuna benzetilebilir (Şekil 16).

**Veri Setinin Organizasyonu ve Kullanımı**  
Veri setlerinin organizasyonu hedef probleme, teknolojik altyapıya ve kullanılacak
programlama diline göre gerçekleştirilmektedir. Yazılan Makine Öğrenmesi kodları, denetimli
öğrenmeye dayalı problemlerde girdi verileri ayrı bir veri seti dosyasında, çıktı verileri ayrı bir
veri seti dosyasından okuyabilmekte ya da geliştirilen uygulama veri setini Makine Öğrenmesi
algoritması için bu yönde parçalayabilmektedir (Benzer şekilde Makine Öğrenmesi süreçleri
de veri seti dosyası üzerinde değişiklikler yapabilmekte ya da yeni bir dosya oluşturarak
problem çözüm aşamalarını destekleyebilmektedir). Denetimsiz öğrenmede ise sadece girdi
verilerini içeren veri setleri kullanılmakta; Makine Öğrenmesi aşamaları sonucunda elde edilen
kümeler (çıktı verileri) veri setlerinde çıktı parametrelerini temsil edecek şekilde
yazılabilmektedir (Böylelikle denetimsiz öğrenmenin bir bakıma sadece girdileri bilinen bir
problem için küme / sınıf bazlı çıktı verileri oluşturduğu anlaşılmaktadır).
Makine Öğrenmesi sistemleri kurulurken, probleme özgü olarak tek bir veri seti
kullanılabileceği gibi, çeşitli veri setlerinin sistemi besleyen bir yaklaşım ile daha gelişmiş
Makine Öğrenmesi modelleri de kurulabilmektedir. Örneğin, bir hastalık teşhisi probleminde
semptomlarla hastalık durumunu belirten spesifik bir veri seti kullanılabilirken, hastaların
farklı tahlil sonuçlarını kullanarak hastalık teşhisinde bulunan, ardından teşhise göre tedavi
planlaması da önerebilen bir sistemde, kurulacak yapı birden fazla veri setinin yine birden fazla
Makine Öğrenmesi algoritmasıyla karşılandığı gelişmiş bir karar destek sistemi olarak
tasarlanabilmektedir. Kimi zaman bir veri setinin tekabül ettiği çıktı verileri (sonuçlar) başka
bir veri setindeki herhangi bir girdi olarak da kullanılabilmektedir. Bu tür kullanım şekilleri
veri setinin “çevrimiçi” dosyalardan ya da “veritabanlarından” okunarak işletilmesi yönünde
desteklenebileceği gibi, kurulan Yapay Zeka sistemi “gerçek zamanlı” yani anlık veriyi
kullanarak arkaplanda veri seti halinde değerlendiren bir mimariye dayandırılarak da
geliştirilebilmektedir. Bu tür yaklaşımlar tamamen problem modelleme ve problem zorluğu ile
veri toplama ve işleme aşamalarında ihtiyaç duyulan mimari yapılanmalarına göre belirlenmek
zorundadır
